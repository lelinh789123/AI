```python
#chon 1999-2010 co bao nhieu huy chuong cac loai cua phap
import pandas as pd
url='http://winterolympicsmedals.com/medals.csv'
data=pd.read_csv(url,sep=',')
data1=data[(data['Year']<2011)&(data['Year']>1998)&(data['NOC']=='FRA')]
data1.groupby('Medal').count()
```

```python
##chon 1999-2010 số huy chương các loại của Pháp ở các môn thể thao
import pandas as pd
url='http://winterolympicsmedals.com/medals.csv'
data=pd.read_csv(url,sep=',')
data1=data[(data['Year']<2011)&(data['Year']>1998)&(data['NOC']=='FRA')]
data1.groupby(['Sport','Medal']).count()
```
